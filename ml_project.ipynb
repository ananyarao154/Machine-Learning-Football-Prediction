step 3) Cleaning our data

```python
df.dtypes #taking a look at each type of data in the column, ml algos can only work with numeric data,so we need to create columns that are numeric and can be used as predictors
```

```python
df['date'] = pd.to_datetime(df['date']) #we are not creating a new column, here we are just using function to kind of convert it into int and then over writing it, confirmation required.
```

```python
df.dtypes
```

step 3) Creating predictors

```python
df['venue_code'] =df['venue'].astype('category').cat.codes    #converting to numeric, .catcodes converts it to integers, basically converting string to category and then category to numbers
```

```python
df #we can see 'Away' = 0 and 'Home' =1
```

```python
df['op_codes'] =df['opponent'].astype('category').cat.codes
```

```python
df #each opponent has their own code
```

```python
df['hour']=df['time'].str.replace(":.+","",regex = True).astype("int") #replacing punctuations so as to make column numeric, in my opinion
```

```python
df #here we see we have created an hour column which is in the form of int
```

```python
df["day_code"] = df["date"].dt.dayofweek
```

```python
df #again there is a new column with int, we created it by using dayofweek
```

```python
df["target"] = (df['result']=='W').astype('int')
#basically here we are converting W = 1(true) and L or D = 0(false)
```

```python
df
```

step 3) Creating our initial ML model

```python
from sklearn.ensemble import RandomForestClassifier #this is a type of ML model that can pick up non-linearities in the data
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split=10,random_state = 1)
#n_estimators is the number of individual decision trees we want to train,higher that number, longer it will take to run but more accurate
#higher the min_samples_split less likely to overfit, but lower accuracy, it is basically the number of samples we want on the leaf of a decision tree
#
```

```python
train = df[df["date"]< '2022-01-01'] #write explanation
```

```python
test = df[df['date']>'2022-01-01'] #need to write explanation
```

```python
predictors = ['venue_code','op_codes','hour','day_code']
```

```python
rf.fit(train[predictors],train['target']) #here we are trying to train our predictors and try to predict target
```

```python
preds = rf.predict(test[predictors]) #we are starting to generate predictions using the predict method on the test data
```

```python
#now determining accuracy
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'],preds) #passing in our actuals and then predictions
```

```python
acc
```

```python
#now checking where or in which situations our accuracy was high and low
#to do this we create a combined df which will combine our actual values and predicted values
combined = pd.DataFrame(dict(actual=test["target"],prediction = preds))
```
